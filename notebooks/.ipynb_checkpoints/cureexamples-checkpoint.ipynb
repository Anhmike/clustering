## Clustering Using REpresentatives (CURE)

Clustering is useful in data mining and statistical data analysis for grouping similar data points together and identifying patterns in distributions.

The majority of clustering algorithms work well for spherical objects, but break down when there are outliers in the dataset. One way to deal with these outliers is to use CURE clustering which is more rebust to datasets with large differences in size or containing outliers. 

In CURE clustering, each cluster is represented by a certain number of representative points. These representative points are chosen from the cluster by taking the the most scattered points in the cluster and then shrinking them towards the centre of the cluster by a fixed amount, known as the compression. This shrinking helps to reduce the effect of outliers.

Below a kd-tree is also deployed in order to store the representative points of each cluster.

### Requirements

The technical dependencies required to run this notebook are as follows:

    - embedPy
    - Matplotlib 2.1.1

### Load CURE

The CURE algorithm is contained in `cure.q` and loads in the script `kdtree.q` to create the kd-tree.

```q
\l ../cure.q
plt:.p.import[`matplotlib]`:pyplot;
```

Below is a dataset with four obvious clusters. Clustering algorithms which do not use representative points would struggle to correctly classify the data given that the clusters mainly share the same centre.

```q
sample:(4000#"F";",")0:`:../data/aadata.csv
plt[`:scatter][;]. flip sample;
plt[`:show][];
```

`.kd.cure.cure` is the function that breaks the function up into the desired cluster groups. The function takes four inputs `.kd.cure.cure[sample;x;y;z]` where `sample` is the data that needs to be clustered, `x` is the amount of representative points, `y` is the compression which is number between 0 and 1 and `z` is the amount of clusters that the data is to be split into. The function returns the indices of the sample grouped into clusters.

```q
show cl:.kd.cure.cure[sample;30;0;4]
```

To check if the clustering is correct, the results can be plotted using `matplotlib.pyplot`, with each cluster represented with a different colour. 

```q
{plt[`:scatter][;]. flip sample[x]}each cl;
plt[`:show][];
```

### Adding compression

The following is an example of how adding the compression can improve clustering when the dataset contains outliers and has varying sizes of clusters.

```q
sample2:(700#"F";",")0:`:../data/fcpstarg.csv
plt[`:scatter][;]. flip sample2;
plt[`:show][];
```

From the above image it is evident that there is a large spread of data points away from the centre which could cause issues if there is a small amount of representative points used.

```q
show cl2:.kd.cure.cure[sample2;5;0;6]
```

```q
{plt[`:scatter][;]. flip sample2[x]}each cl2;
plt[`:show][];
```

The above image shows how the centre cluster and its surrounding ring are not classified correctly. This can be rectified by adding a compressing which moves the representative points of each cluster towards its centre to deal with varying shapes and outliers.

```q
show clcomp:.kd.cure.cure[sample2;5;0.1;6]
```

```q
{plt[`:scatter][;]. flip sample2[x]}each clcomp;
plt[`:show][];
```

By adding a compression of 0.1, the datapoints were correctly clusters.
