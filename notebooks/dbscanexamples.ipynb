## DBSCAN Clustering

Density-based spatial clustering of applications with noise (DBSCAN) is a density-based clustering algorithm which groups data points together that are closely packed together in areas of high-density. Any points in low-density regions are seen as outliers.

The algorithm requires 3 inputs:
- data: matrix of data points
- minimum points: minimum number of points required in a given neighbourhood for it to be classified as a cluster
- epsilon radius: distance from each point, within which, points are defined as being in the same cluster assuming that the minimum number of points is met

Unlike other clustering algorithms which require the user to input the number of cluster in the dataset, DBSCAN will decide how many clusters are in the dataset based on how many high-density regions are present. The benefit of this method is that it is fast to run and deals well with outliers. However, the algorithm stuggles with datasets containing varying densities.
